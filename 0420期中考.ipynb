```
install.packages("devtools")
devtools::install_github('cttobin/ggthemr')
```

```
library(ggthemr)
ggthemr("dust")
NBA1516 %>%
 count(Position) %>%
 ggplot(aes(x=fct_reorder(Position,n), y=n , fill=Position))+
 geom_col()+
 geom_text(aes(label=n),vjust=-0.25)+
 labs(title="Position-Quentity",x="Position",y="Quantity")+
 theme(legend.position = "none")
```

```
NBA1516  %>%
  mutate(Position=fct_reorder(Team,TotalPoints,median,na.rm=TRUE)) %>%
  ggplot(aes(Team,TotalPoints))+
  geom_point()+
  geom_boxplot()+
coord_flip()
```

```
install.packages("ggridges")
```

```
library(ggridges)
NBA1516  %>%
ggplot(aes(
  x=TotalPoints,
  y=Position,
  fill=Position
))+
geom_density_ridges()
```
