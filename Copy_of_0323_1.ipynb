```
#安裝
install.packages("ggplot2")
install.packages("SportsAnalytics")
```

```
#載入
library(ggplot2)
library(SportsAnalytics)
```

```
#讀入
NBA1516<-fetch_NBAPlayerStatistics("15-16")
```

```
str(NBA1516)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1516)
```

```
qplot(ThreesAttempted,TotalPoints,data = NBA1516)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1516,color=Position)
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1516,color=Team)
```

```
qplot(ThreesAttempted,TotalPoints,data = NBA1516,color=Position)
qplot(ThreesAttempted,TotalPoints,data = NBA1516,color=Team)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1516,geom=c("point","smooth"))
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1516,color=Team,geom=c("point","smooth"))
qplot(ThreesAttempted,TotalPoints,data = NBA1516,color=Team,geom=c("point","smooth"))
```

```
qplot(TotalPoints,data=NBA1516,fill=Position)
qplot(TotalRebounds,data=NBA1516,fill=Position)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1516,facets=Position~.)
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1516,facets=.~Position)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1516,facets=Position~. ,binwidth=10)
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1516,facets=.~Position ,binwidth=100)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1516,facets=Team~. ,binwidth=10)
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1516,facets=.~Team ,binwidth=100)
```
